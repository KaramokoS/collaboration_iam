### Opération sur de colonnes

Nous pouvons également créer de nouvelles colonnes ou modifier celles existantes.

```python
ventes['Revenu_par_Age'] = ventes['Revenu'] / ventes['Age_Client']

ventes['Revenu_par_Age'].head()
```

```python
ventes['Revenu_par_Age'].plot(kind='density', figsize=(14,6))
```

```python
ventes['Revenu_par_Age'].plot(kind='hist', figsize=(14,6))
```

#### Ajouter et calculer une nouvelle colonne Calculated_Cost
Utilisons la formule ci-dessous:
- $Cout\_Calculé = Quantité*Cout\_Unitaire$

```python
ventes['Cout_Calculé'] = ventes['Quantité'] * ventes['Cout_Unitaire']

ventes['Cout_Calculé'].head()
```

```python
(ventes['Cout_Calculé'] != ventes['Coût']).sum()
```

Nous pouvons voir la relation entre le coût et le profit à l’aide d’un nuage de points :

```python
ventes.plot(kind='scatter', x='Cout_Calculé', y='Profit', figsize=(6,6))
```

#### Ajouter et calculer une nouvelle colonne Calculated_Revenue
Utilisons la formule ci-dessous:
- $Revenu\_Calculé = Coût + Profit$

```python
ventes['Revenu_Calculé'] = ventes['Coût'] + ventes['Profit']

ventes['Revenu_Calculé'].head()
```

```python
(ventes['Revenu_Calculé'] != ventes['Revenu']).sum()
```

```python
ventes.head()
```

```python
ventes['Revenu'].plot(kind='hist', bins=100, figsize=(14,6))
```
